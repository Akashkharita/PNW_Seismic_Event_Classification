```python
# Save to a file
with open('../results/event_report_tsf.pkl', 'rb') as pickle_file:
     event_report_tsfel = pickle.load(pickle_file)


with open('../results/event_report_phy.pkl', 'rb') as pickle_file:
     event_report_physical = pickle.load(pickle_file)
        
        
with open('../results/event_report_phy_tsf_man.pkl', 'rb') as pickle_file:
     event_report_all = pickle.load(pickle_file)
        
with open('../results/event_report_phy_tsf_man.pkl', 'rb') as pickle_file:
     event_report_all_manual = pickle.load(pickle_file)
        
        
        
with open('../results/event_report_scatnet_features_without_manual_features.pkl', 'rb') as pickle_file:
     event_report_scatnet = pickle.load(pickle_file)
        
        
with open('../results/event_report_archtime_do.pkl', 'rb') as pickle_file:
     event_report_archtime = pickle.load(pickle_file)
        
with open('../results/event_report_seismicnet_do.pkl', 'rb') as pickle_file:
     event_report_seismicnet = pickle.load(pickle_file)
        
with open('../results/event_report_wavedecompnet_do.pkl', 'rb') as pickle_file:
     event_report_wavedecompnet = pickle.load(pickle_file)
        
with open('../results/event_report_seismiccnn_batch_do.pkl', 'rb') as pickle_file:
     event_report_seismiccnn = pickle.load(pickle_file)
```

## Trace wise report

```python
plt.rcParams['ytick.labelsize'] = 20  # Font size for ytick labels


# Sample data (replace with your actual data)
classes = ['Earthquake', 'Explosion', 'Noise', 'Surface']
variables = ['ScatNet(ML)', 'Physical(ML)', 'TSFEL(ML)', 'TSFEL+Physical(ML)',
             'TSFEL + Physical + Manual(ML)', 'SeismicCNN(DL)',
             'WaveDecompNet(DL)', 'Archtime(DL)', 'SeismicNet(DL)']

# Creating a 2D array of values for each class and variable (Assuming these are defined elsewhere in your code)
# Set rc parameters for font size



# Creating a 2D array of values for each class and variable
values_precision = np.array([
    pd.DataFrame(trace_report_scatnet).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_physical).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_tsfel).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_all).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_all_manual).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_seismiccnn).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_wavedecompnet).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_archtime).loc['precision'].values[0:4],
    pd.DataFrame(trace_report_seismicnet).loc['precision'].values[0:4]
])

values_recall = np.array([
    pd.DataFrame(trace_report_scatnet).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_physical).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_tsfel).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_all).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_all_manual).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_seismiccnn).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_wavedecompnet).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_archtime).loc['recall'].values[0:4],
    pd.DataFrame(trace_report_seismicnet).loc['recall'].values[0:4]
])

values_accuracy = np.array([
    [pd.DataFrame(trace_report_scatnet)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_physical)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_tsfel)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_all)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_all_manual)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_seismiccnn)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_wavedecompnet)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_archtime)['accuracy'].values[0]],
    [pd.DataFrame(trace_report_seismicnet)['accuracy'].values[0]]
])




# Set up the figure and axes
fig, axes = plt.subplots(3, 1, figsize=(18, 18))

# Define parameters for bar plots
bar_width =  1 / (len(variables) + 1)
bar_positions = np.arange(len(classes))
acc_bar_positions = np.arange(len(variables))
colors = ['#2e86c1', '#27ae60', '#e74c3c', '#f39c12', '#8e44ad', '#3498db', '#c0392b', '#f39c12', '#16a085']

ml_models = ['ScatNet(ML)', 'Physical(ML)', 'TSFEL(ML)', 'TSFEL+Physical(ML)', 'TSFEL + Physical + Manual(ML)']

# Define labels for subplots
sub_labels = ['(a)', '(b)', '(c)']

# Iterate over each subplot
for idx, (ax, values, ylabel, title, label) in enumerate(zip(axes, [values_precision, values_recall, values_accuracy],
                                     ['Precision Values', 'Recall Values', 'Accuracy Values'],
                                     ['Precision',
                                      'Recall',
                                      'Accuracy'], sub_labels)):

    for i, var in enumerate(variables):
        edge_color = 'black' if var in ml_models else None
        linewidth = 2 if var in ml_models else 0  # Increase thickness for ML models
        pattern = '.O' if var in ml_models else ''  # Add pattern for ML models
        if title == 'Accuracy':
            bar_width = 0.3
            ax.bar(acc_bar_positions[i], values[i, 0], width=bar_width, label=var, color=colors[i], edgecolor=edge_color, linewidth=linewidth, hatch = pattern)
        else:
            ax.bar(bar_positions + i * bar_width, values[i, :], width=bar_width, label=var, color=colors[i], edgecolor=edge_color, linewidth=linewidth, hatch = pattern)

    ax.set_xticks(bar_positions + (len(variables) - 1) * bar_width / 2)
    ax.set_xticklabels(classes, fontsize=30)
    ax.set_xlabel('Class', fontsize=20)
    ax.set_ylabel(ylabel, fontsize=30)
    ax.set_ylim(0.1, 1.05)

    for i, var in enumerate(variables):
        for j, value in enumerate(values[i, :]):
            if title != 'Accuracy':
                ax.text(bar_positions[j] + i * bar_width, value + 0.01, f'{value:.2f}', ha='center', va='bottom',
                        fontsize= 12)
            else:
                ax.text(acc_bar_positions[i], value + 0.01, f'{value:.2f}', ha='center', va='bottom',
                        fontsize=20)

    ax.set_title(title, fontsize=25)

    if title == 'Accuracy':
        ax.set_xticks([])
        ax.set_xticklabels([])
        ax.set_xlabel('', fontsize=30)
        ax.set_ylim(0.7, 1.0)

    # Add subplot label
    ax.text(0.01, 0.9, label, transform=ax.transAxes, fontsize=20)

fig.suptitle("Trace wise results", fontsize = 30)
plt.legend(fontsize=20, ncol=3, loc = 'lower right')
plt.tight_layout()
plt.savefig('../plots/trace_wise_model_performance_comparisons.png')
plt.show()
```
